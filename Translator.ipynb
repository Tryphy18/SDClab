```python
from transformers import pipeline
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.llms.huggingface_pipeline import HuggingFacePipeline

# ✅ Load Hugging Face translation pipeline (English to Romance languages like Spanish)
translator_pipeline = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ROMANCE")

# ⛓️ Wrap into LangChain LLM
llm = HuggingFacePipeline(pipeline=translator_pipeline)

# 🔤 Prompt Template
prompt = PromptTemplate(
    input_variables=["text"],
    template="Translate the following English sentence into a Romance language (like Spanish, French, or Italian):\n\n{text}"
)

# 🔁 LangChain Chain
translator_chain = LLMChain(llm=llm, prompt=prompt)

# 📝 Input text to translate
text_to_translate = "I love learning new languages and exploring different cultures."

# 🚀 Translate
translated_output = translator_chain.run(text=text_to_translate)

# 📤 Output
print("Original Text:", text_to_translate)
print("Translated Text:", translated_output)
```
