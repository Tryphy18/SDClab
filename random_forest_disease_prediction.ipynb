```python
import gradio as gr
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Sample Data: [Fever, Cough, Fatigue] - 1 = Present, 0 = Absent
X = np.array([
    [1, 1, 1],  # Fever, Cough, Fatigue -> Flu
    [1, 0, 1],  # Fever, No Cough, Fatigue -> Flu
    [0, 1, 0],  # No Fever, Cough, No Fatigue -> Flu
    [0, 0, 1],  # No Fever, No Cough, Fatigue -> No Flu
    [0, 0, 0],  # No Symptoms -> No Flu
    [1, 1, 0],  # Fever, Cough, No Fatigue -> Flu
    [1, 0, 0],  # Fever, No Cough, No Fatigue -> Flu
    [0, 1, 1]   # No Fever, Cough, Fatigue -> Flu
])

# Labels: 1 = Flu, 0 = No Flu
y = np.array([1, 1, 1, 0, 0, 1, 1, 1])

# Train the Random Forest model
model = RandomForestClassifier(n_estimators=100)
model.fit(X, y)

# Prediction function
def predict_disease(fever, cough, fatigue):
    symptoms = np.array([[fever, cough, fatigue]])
    prediction = model.predict(symptoms)[0]
    return "ðŸ¦  Flu" if prediction == 1 else "ðŸ˜Š No Flu"

# Gradio interface
gr.Interface(
    fn=predict_disease,
    inputs=[
        gr.Radio([1, 0], label="Fever (1 = Yes, 0 = No)"),
        gr.Radio([1, 0], label="Cough (1 = Yes, 0 = No)"),
        gr.Radio([1, 0], label="Fatigue (1 = Yes, 0 = No)")
    ],
    outputs="text",
    title="Disease Prediction App",
    description="Predict if you have Flu based on symptoms (Fever, Cough, Fatigue)."
).launch()
```
